```python
from sklearn.datasets import load_iris
import numpy as np
import pandas as pd
np.random.seed(0)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
iris=load_iris()
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df.head()
```

```python
df['species']=pd.Categorical.from_codes(iris.target,iris.target_names)
df.head()
```

```python
df['is_train']=np.random.uniform(0,1,len(df))<=.75 #generating random number between 0 and 1 to all rows-len(df) if its less than 
                                                    #0.75 is_train is true else flase 
df.head()
```

```python
train=df[df['is_train']==True]
test=df[df['is_train']==False]
print("length of train ",len(train))
print("length of test ",len(test))
```

```python
features=df.columns[:4]
features
```

```python
#converting each species name into digits
y=pd.factorize(train['species'])[0]
y  #0,1,2-types of species
```

```python
clf=RandomForestClassifier(n_jobs=2,random_state=0)  # n_jobs number of priorities
clf.fit(train[features],y)
```

```python

clf.predict(test[features])
```

```python

clf.predict_proba(test[features])[10:20]
```

```python
pred=iris.target_names[clf.predict(test[features])]
```
